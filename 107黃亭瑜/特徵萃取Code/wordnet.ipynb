```python
f = open('Formspring/BadWordList.txt', 'r').read() #str
file = f.split(',')  # str to list,以','切割
```

```python
poslist = [wn.NOUN, wn.VERB, wn.ADJ, wn.ADV]
wordlist = []
count = 0
for i in file:
    print(i)
    synset = wn.synsets(i,pos = poslist)
#     if(len(synset)!= 0):    #計算有幾個字有擴增=104
#         count+=1
    for j in range(len(synset)):
#         print(synset[j].name())
        print('synset:',wn.synset(synset[j].name()).lemma_names())
    print('***********')
# print(count)
```

```python
poslist = [wn.NOUN, wn.VERB, wn.ADJ, wn.ADV]
badlist = []
count = 0
for i in file:
#     print(i)
    badlist.append(i)
    synset = wn.synsets(i,pos = poslist)
#     if(len(synset)!= 0):    #計算有幾個字有擴增=104
#         count+=1
    for j in range(len(synset)):
#         print(synset[j].name())
        word = wn.synset(synset[j].name()).lemma_names()
        print('synset:',word)
        for k in range(len(word)):
            badlist.append(word[k])
setbadlist = list(set(badlist))  #總共894個 
setbadlist.sort(key=badlist.index)
print(setbadlist)
```

```python
num = 0
exl = []
for a in setbadlist:
    num += 1
    countnum = 0
    if a in file:
        countnum = 1
    exl.append([])
    exl[num-1].append(str(a))
    exl[num-1].append(countnum)
```

```python
title = ['Word', 'BadWord'] #, 'Num Expanded'
with open('Myspace/badexp4.csv', 'w', encoding='UTF-8', newline='') as csvfile:
  # 以空白分隔欄位，建立 CSV 檔寫入器
    writer = csv.writer(csvfile)
    writer.writerow(title)
    writer.writerows(exl)
```
